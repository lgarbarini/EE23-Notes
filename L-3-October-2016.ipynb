# In Class Example

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import time
from IPython.core.display import clear_output,display

f,ax = plt.subplots(nrows=1,ncols=1)
plt.xlim(-7,7)
plt.ylim(-3,3)
solutions = []
ts = np.linspace(-7,7,100)
ns = range(1,100)
for n in ns:
    time.sleep(0.2)
    sum = 0
    temp = []
    for t in range(len(ts)):
        
        fx = (((-1)**(n+1)/n)*2*np.sin(n*ts[t]))
        if n > 1 :
            fx += solutions[n-2][t]
        temp.append(fx)
    solutions.append(temp)
    ax.plot(ts, temp)
    clear_output(wait=True)
    display(f)
    
plt.close()
```

# Square Wave

```python
f,ax = plt.subplots(nrows=1,ncols=1)
plt.xlim(-1,1)
plt.ylim(-1,1)
solutions = []
ts = np.linspace(-1,1,100)
ns = range(1,100)
for n in ns:
    time.sleep(0.1)
    sum = 0
    temp = []
    for t in range(len(ts)):
        
        fx = (np.sin(2*np.pi*(2*n-1)*ts[t])/(2*n-1))
        if n > 1 :
            fx += solutions[n-2][t]
        temp.append(fx)
    solutions.append(temp)
    ax.plot(ts, temp)
    clear_output(wait=True)
    display(f)
    
plt.close()
```
