# In Class Example

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import time
from IPython.core.display import clear_output,display

f,ax = plt.subplots(nrows=1,ncols=1)
plt.xlim(-7,7)
plt.ylim(-3,3)
solutions = []
fx = 0
ts = np.linspace(-7,7,1000)
ns = range(1,20)
for n in ns:
    time.sleep(0.1)
    temp = []
    for t in range(len(ts)):
        fx = (((-1)**(n+1))/n)*2*np.sin(n*ts[t])
        if n > 1 :
            fx += solutions[n-2][t]
        temp.append(fx)
    solutions.append(temp)
    ax.plot(ts, temp)
    clear_output(wait=True)
    display(f)
    
plt.close()
plt.show()
```

# In Class Example -m to positive m

```python
m = 20 ; # number of terms (equally from the postive and
# negative side) to add up in the Fourier Series
t = np.linspace(-7,7,10000); # time support of the signal
y = np.zeros(10000) # Approximation = S_m(x(t))
f,ax = plt.subplots(nrows=1,ncols=1)
plt.xlim(-7,7)
plt.ylim(-4,4)
plt.xlabel('time');
for n in range(-m,m): # total terms 2*m + 1 in complex Fourier series
    if n != 0:
        y = y + (((-1)**(n+1)/n)*2*np.sin(n*t))
        time.sleep(0.1)
        ax.plot(t,y)
        clear_output(wait=True)
        display(f)
plt.close()
```

# Square Wave

```python
f,ax = plt.subplots(nrows=1,ncols=1)
plt.xlim(-1,1)
plt.ylim(-1,1)
solutions = []
ts = np.linspace(-1,1,10000)
ns = range(-20,20)
for n in ns:
    time.sleep(0.1)
    sum = 0
    temp = []
    for t in range(len(ts)):
        fx = (np.sin(2*np.pi*(2*n-1)*ts[t])/(2*n-1))
        if n > -19 :
            fx += solutions[n+19][t]
        temp.append(fx)
    solutions.append(temp)
    ax.plot(ts, temp)
    clear_output(wait=True)
    display(f)
    
plt.close()
```
